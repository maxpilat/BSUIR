```python
import pandas as pd
import re
dataset = pd.read_csv("spotify_dataset.csv")

dataset
```

```python
dataset = dataset.drop(["Peak Position", "Peak Streams"], axis=1)

dataset
```

```python
def extract_number(value):
    if isinstance(value, int):
        return value
    match = re.search(r'\d+',value)
    if match:
        return int(match.group())
    else:
        return 0

for index, row in dataset.iterrows():
    dataset.loc[index, "Peak Position (xTimes)"] = extract_number(dataset["Peak Position (xTimes)"][index])
    
dataset
    
```

```python
dataset["Artist Name (factorized)"] = pd.factorize(dataset["Artist Name"])[0]

dataset
```

```python
dataset["Total Streams"] = (dataset["Total Streams"] - dataset["Total Streams"].min()) / (dataset["Total Streams"].max() - dataset["Total Streams"].min())
dataset["Total Streams"] = (dataset["Total Streams"] - dataset["Total Streams"].mean()) / dataset["Total Streams"].std()
dataset.rename(columns={'Total Streams': 'Total Streams (Normalized and Standardized'}, inplace=True)

dataset
```

```python
dataset.to_csv("spotify_dataset(treated).csv")
```
